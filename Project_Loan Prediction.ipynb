```python
#from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
models=[]
models.append(("logreg",LogisticRegression()))
models.append(("tree",DecisionTreeClassifier()))
models.append(("rfc",RandomForestClassifier()))
#models.append(("lda",LinearDiscriminantAnalysis()))
models.append(("svc",SVC()))
models.append(("knn",KNeighborsClassifier()))
models.append(("nb",GaussianNB()))
```

```python
models
```

```python
seed=7
scoring='accuracy'
```

```python
from sklearn.model_selection import KFold 
from sklearn.model_selection import cross_val_score
result=[]
names=[]
```

```python
for name,model in models:
    print(model)
    kfold=KFold(n_splits=10)
    cv_result=cross_val_score(model,train_X,train_y,cv=kfold,scoring=scoring)
    result.append(cv_result)
    names.append(name)
    print(" %f %f" % (cv_result.mean(),cv_result.std()))
    print("\n\n")
    #print(cv_result)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
lgr=LogisticRegression()
lgr.fit(train_X,train_y)
lgrpred=lgr.predict(test_X)
print(accuracy_score(test_y,lgrpred))
print(confusion_matrix(test_y,lgrpred))
print(classification_report(test_y,lgrpred))
```

```python
precisionlgr = precision_score(test_y, lgrpred, average="weighted")
recalllgr = recall_score(test_y, lgrpred, average="weighted")
f1lgr = f1_score(test_y, lgrpred, average="weighted")
print("Logistic Regression")
print("Precison is :",precisionlgr)
print("Recall is :",recalllgr)
print("f-1 score is :",f1lgr)
print("accuracy: ",accuracy_score(test_y,lgrpred))
```

```python
dtc=DecisionTreeClassifier(max_depth=None)
dtc.fit(train_X,train_y)
dtcpred=dtc.predict(test_X)
accudtc=accuracy_score(test_y,dtcpred)
print("The accuracy for Decision Tree is : ",accuracy_score(test_y,dtcpred))
print("The Confusion Matrix  is : \n",confusion_matrix(test_y,dtcpred))
print(classification_report(test_y,dtcpred))
```
